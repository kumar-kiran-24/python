```python
words=["eat","eating","eaten","writes","writr","writting","programmig","programs"]
```

# porteStemmer


```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
```

```python
for word in words:
    print(word+"----"+stemming.stem(word))
```

# regexpStemmer class
easily implment regular expression stemmer alogrothms

```python
from nltk.stem import RegexpStemmer
reg=RegexpStemmer("ing$$|s$|es$|abel$",min=4)
```

```python
reg.stem("eating") 
```

# snowball stemmer

```python
from nltk.stem import SnowballStemmer
s_b=SnowballStemmer("english")
```

```python
for word in words:
    print(word+"----"+s_b.stem(word))
```
